```python
#constantes
c=1
T=0.15

n_points=200
iteraciones=1000

U_initial=zeros([n_points,1],dtype=float)
space = linspace(0.0,2.0,n_points)
u_initial = 3*sin(math.pi*space)
plot(space,u_initial,"yo", label="Condicion inicial")
legend ()
```

```python
delta_x=space[1]-space[0]
c=1.0
delta_t= T/iteraciones
beta= c*delta_t/delta_x
print delta_t
```

```python
#Condiciones fijas
u_initial[0]=0
u_initial[-1]=0

inicial=3*sin(pi*space)
```

```python
for j in range(iteraciones):
    
    if (j== 1):
       plot(space,u_initial) 
        
    if (j== 200):
       plot(space,u_initial)
        
    if (j== 400):
       plot(space,u_initial)
        
    if (j== 600):
       plot(space,u_initial)
        
    if (j== 800):
       plot(space,u_initial)
        
    if (j== 1000):
       plot(space,u_initial)
        
    u_past=u_initial.copy()
    
    for i in range(1,n_points-1):
        
         u_initial[i]=u_past[i]-(beta*0.5)*( u_past[i+1]**2  -  u_past[i-1]**2 )
        
    
    
u_future=u_initial.copy()  

```

```python
plot(space,u_future,"r",label="final")
plot(space,inicial,label="inicial")
legend()
```
