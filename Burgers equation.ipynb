```python
#definicion de condiciones iniciales
c=1
n_points = 200

#iteraciones en el tiempo
iteraciones=1000
#tiempo asignado
T=0,15

x = linspace(0.0,2.0,n_points)
u_initial = 3*sin(math.pi*x)
```

```python
plot(x,u_initial)
```

```python
#condiciones iniciales
delta_x=x[1]-x[0]
c=1.0
delta_t= 0.0015
beta= c*delta_t/delta_x


#beta debe satisfacer la condicion beta<1
print beta


u_initial[0] = 0.0
u_initial[1] = 0.0

#u_future = uj+1
u_future = zeros(n_points)
u_future[0] = 0.0
u_future[1] = 0.0


for i in range(1,n_points-1):
    u_future[i] = u_initial[i] -(beta*0.5)*(u_initial[i+1]**2-u_initial[i-1])

#create a new variable to hold the previous value
u_past = u_initial.copy()
#create a new variable to hold the present value
u_present = u_future.copy()

```

```python
plot(x, u_initial)
plot(x, u_past)
plot(x, u_present)
plt.xlabel('x')
plt.ylabel('u')
```
