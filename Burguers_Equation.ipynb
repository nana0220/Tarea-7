```python
%pylab inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
#constantes
c=1
T=0.15

n_points=200
iteraciones=1000

delta_x=2.0/n_points
delta_t=T/iteraciones

beta=c*delta_t/delta_x
print beta
```

```python
#definiendo arrays:
u=zeros([n_points,1],dtype=float)
u_initial=zeros([n_points,1],dtype=float)
space=linspace(0,2,n_points)
```

```python
u=3*sin(pi*space)
u_initial=3*sin(pi*space)
```

```python
plot(space,u)
```

```python
u[0]=0
u[-1]=0
```

### Solucion 1

```python
for j in range(iteraciones):
    plot(space,u)
    u_past=u.copy()
    
    for i in range (1,n_points-1):
        
         u[i]= u_past[i]-(beta*0.5)*( u_past[i+1]**2 - u_past[i-1]**2 )
            

u_future=u.copy()
            
            
```

```python
plot(space,u_future,"r",label="final")
plot(space,u_initial,"g",label="inicial")
legend()
```
